Training the model


```python
from ultralytics import YOLO

model = YOLO('yolov8m.pt')


model.train(
    data='plastic.yaml', 
    epochs=80, 
    imgsz=(1280, 720),
    batch=5,
    optimizer="Adam",
    lr0=1e-3,
    seed=63,
    device=0,
)
```

Extracting GPS from the give data(Images)


```python
from PIL import Image #PIL: Pillow package to read metadata from image
from PIL.ExifTags import TAGS
import os

gps_lst = []
def get_gps_data(image_path):
    image = Image.open(image_path)
    exif_data = image._getexif()
    
    if exif_data is not None:
        for tag_id, value in exif_data.items():
            tag_name = TAGS.get(tag_id, tag_id)
            if tag_name == 'GPSInfo':
                # Extract GPS data
                gps_data = {}
                for key in value.keys():
                    sub_tag_name = TAGS.get(key, key)
                    gps_data[sub_tag_name] = value[key]
                return gps_data

    # No GPS data found
    return None
directory = r'datasets\images\Predict'
for filename in os.listdir(directory):
    f = os.path.join(directory, filename)
    # checking if it is a file
    if os.path.isfile(f):
        image_path = f
        gps_data = get_gps_data(image_path)

        if gps_data is not None:
            print("GPS Data:",filename)
            gps=[]
            gps_cords = {}
            for key, value in gps_data.items():
                if(key in [2,4]):
                    gps.append(value[0])
                    gps.append(value[1])

                elif(key in ['InteropIndex',3]):
                    gps.append(value)
            print(gps)
            gps_lst.append(gps)
            

        else:
            print("No GPS data found.")
```

Creating CSV file(Organizing)


```python
import csv
import os
import pandas as pd
import tensorflow as tf


#creating gps link
def location_link(lat_deg, lat_min, lat_sec, lon_deg, lon_min, lon_sec):
    base_url = "https://www.google.com/maps/search/?api=1&query="
    latitude = f"{lat_min}°{lat_sec}'{lat_deg}"
    longitude = f"{lon_min}°{lon_sec}'{lon_deg}"
    coordinates = f"{latitude},{longitude}"
    return base_url + coordinates

#reading predicted bbox data
def pred_bbox_data(pred_label_path,labels):

    pred_count=0
    pred_lab = False
    pred_path = os.path.join(pred_label_path,labels)
    pred_file = open(pred_path,"r")
    pred_data = pred_file.read()
    pred_bbox_data = eval(pred_data)
    return pred_bbox_data

#reading actual bbox data
def actu_bbox_data(actu_label_path,labels):
    actu_path = os.path.join(actu_label_path,labels)
    actu_file = open(actu_path,"r")
    actu_data = actu_file.read()
    data = actu_data.replace(' ',",")
    data_list = data.split('\n')
    return data_list


file = open("data.csv","w") #creating csv file
writer = csv.writer(file)
    
    # Create a new row as a list
new_row = ['IMG_ID', 'PRED_LAB', 'ACTUAL_CT','PRED_CT','CT_ERROR',r'%ERROR','mAP','GPS_CORDS','GEO_Tag_URL'] #adding header data to csv
    
    # Write the new row to the CSV file
writer.writerow(new_row)


gps_ind = 0 #for traversing over gps list


actu_label_path = r'datasets\labels\predict'
pred_label_path = r'predicted\labels'
for labels in os.listdir(pred_label_path):
    
    emp_lst = [] #empty list to append data to rows
    pred_count=0 #for counting no. of plastics

    emp_lst.append(os.path.splitext(labels)[0])

    for prd_index in range(len(pred_bbox_data(pred_label_path,labels))):
        id=pred_bbox_data(pred_label_path,labels)[prd_index][5]
        mAp_data = pred_bbox_data(pred_label_path,labels)[prd_index][4]
        if(id==0.0):
            pred_count+=1
            mAp_data+=mAp_data
            pred_lab = True
    if(pred_lab==True):

        emp_lst.append('YES')
        

        actu_count = len(actu_bbox_data(actu_label_path,labels))
        emp_lst.append(actu_count)  #actual count
        emp_lst.append(pred_count) #predicted count

        ct_error = actu_count - pred_count
        emp_lst.append(ct_error) #ct_error

        error = float(ct_error)/float(actu_count)
        emp_lst.append(error)   #error

        emp_lst.append(metric.box.map50) #appending map value obatined from validation of model


        #modifying gps list to usable form
        gps_link = gps_lst[gps_ind]

        emp_lst.append(gps_link)

        lat_deg = gps_link[0]
        lat_min = int(gps_link[1])
        lat_sec = gps_link[2]
        lon_deg = gps_link[3]
        lon_min = int(gps_link[4])
        lon_sec = gps_link[5]

        loc_link = location_link(lat_deg, lat_min, lat_sec, lon_deg, lon_min, lon_sec)
        print(loc_link)

        link = f'=HYPERLINK("{loc_link}","location")'
        emp_lst.append(link)  #appending gps link
        gps_ind+=1

    else:
        emp_lst.append('NO')




    writer.writerow(emp_lst) #writing all the data to csv file 




        
        
    
```
